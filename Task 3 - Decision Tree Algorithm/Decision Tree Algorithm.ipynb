# Decision Tree Classifier for the Iris Dataset
### Task 3 - To Explore Decision Tree Algorithm

In this task, we will create a Decision Tree classifier for the Iris Dataset and visualize it graphically. We will also evaluate our classifier's performance.

### Importing Libraries

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

### Loading the Data

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data

iris_data = pd.DataFrame(X, columns = iris.feature_names)

y = iris.target
```

```python
iris_data.head()
```

```python
y
```

```python
print(iris.target_names)
```

Setosa is coded as 0, Versicolor as 1, and Virginica as 2.

### Splitting the data into Training and Test Sets

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4)
```

### Training the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

### Visualising the Decision Tree

```python
from sklearn.tree import plot_tree

fig = plt.figure(figsize = (20,10))
plot_tree(classifier, feature_names = iris.feature_names, class_names=iris.target_names, filled = True, rounded = True)
plt.show()

fig.savefig('DecisionTree.png')
```

### Making Predictions

```python
y_pred = classifier.predict(X_test)
y_pred
```

### Evaluating the Model

```python
# Replacing the coded classes with the actual class names
# Setosa: 0, Versicolor: 1, Virginica: 2
# iris.target_names = ['setosa' 'versicolor' 'virginica']
y_test = iris.target_names[y_test]
y_pred = iris.target_names[y_pred]

df = pd.DataFrame(np.column_stack((X_test, y_test, y_pred)), columns = iris.feature_names + ['Actual Class', 'Predicted Class'])

def highlight(s, column):
    if s['Actual Class'] == s['Predicted Class']:
        return ['background-color: lightgreen']*6
    else:
        return ['background-color: salmon']*6
    
df.style.apply(highlight, column=['Actual Class', 'Predicted Class'], axis=1)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

Thus, we can say that our classification rate is **96.67%**.
